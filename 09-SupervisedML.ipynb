# 09 - Supervised ML

Learning goals:

* Introduction to Supervised ML
  * Training dataset includes target variables
  * Prediction vs inference
    * Machine Learning (CS) focus is prediction
    * Statistical learning includes inference
    * The distinction is disciplinary, not fundamental
  * Supervised vs unsupervised ML
    * no target variables with unsupervised ML
    * clustering (unsupervised -- next week)
    * dimensionality reduction (unsupervised -- next week)
* Building classification models
  * Regression vs Classification
    * Regression involves modeling quantitative target variable
    * Classification involves modeling categorical variables
  * Types of models
    * Gaussian Naive Bayes
    * Logistic Regression
    * Others: decision trees, boosting, bagging, support vector machines
    * Linear vs nonlinear (kernels)
  * [Classifier comparison](https://scikit-learn.org/stable/auto_examples/classification/plot_classifier_comparison.html) -- scikit-learn.org
* Dealing with class imbalance
  * Alternative metrics for model performance
  * Compensating for sampling bias
